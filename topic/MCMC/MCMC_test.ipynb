A great target function for practicing the MCMC (Markov Chain Monte Carlo) algorithm with oscillation features could be a combination of a Gaussian distribution with a sinusoidal modulation. Such a function offers both the simplicity of understanding (due to the Gaussian component) and the complexity needed for practice (due to the oscillations). Here’s an example function you can use:

$p(x) = \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right) \cdot (1 + \sin(\omega x))$

Where:
- $\mu$ is the mean of the Gaussian distribution.
- $\sigma$ is the standard deviation of the Gaussian distribution.
- $\omega$ controls the frequency of the sinusoidal oscillations.

For instance, you could set:
- $\mu = 0$ 
- $\sigma = 2$
- $\omega = 2\pi$, which would give oscillations every unit interval.

This function is positive and integrates to a finite value, making it suitable as a probability distribution after proper normalization (although in practice for MCMC, normalization is not strictly necessary as it often cancels out when calculating ratios).

Let’s plot this function to visualize it:

Here's the plot of the target function $p(x) = \exp\left(-\frac{(x-0)^2}{8}\right) \cdot (1 + \sin(2\pi x))$ for practicing MCMC. This function includes both a Gaussian bell curve centered at zero and sinusoidal oscillations, making it an interesting case for MCMC simulations due to its multiple modes and varying density. This function should provide a good challenge and learning experience for students. 

```python
# Define parameters for the target function
mu = 0
sigma = 2
omega = 2 * np.pi

# Define the target function
x = np.linspace(-10, 10, 400)
p_x = np.exp(-((x - mu)**2) / (2 * sigma**2)) * (1 + np.sin(omega * x))

# Normalize p_x to make it a proper probability density function
p_x /= np.trapz(p_x, x)

# Plot the target function
plt.figure(figsize=(10, 5))
plt.plot(x, p_x, label='p(x) = exp(-((x-0)^2)/8) * (1 + sin(2πx))')
plt.title('Target Function for MCMC Practice with Oscillation')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.grid(True)
plt.legend()
plt.show()
```

This function includes both a Gaussian bell curve centered at zero and sinusoidal oscillations, making it an interesting case for MCMC simulations due to its multiple modes and varying density.
